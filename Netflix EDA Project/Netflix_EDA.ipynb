**Movie subset**

```python
#subset that only include movie

netflix_movie = netflix[netflix["type"]== "Movie"]
netflix_movie
```

```python
#getting the count director in the dataset

netflix_movie["director"].value_counts()
```

```python
#split the movie director to only include 1 name

movie_directors = netflix_movie["director"].str.split(",", expand = True).stack()
movie_directors
```

```python
#convert to a dataframe

movie_directors = pd.DataFrame(movie_directors)
movie_directors.head()
```

```python
movie_directors.columns = ["Directors"]
```

```python
#movie_directors = movie_directors[movie_directors["Directors"] != "Not specified"]
```

```python
#only selecting the known directors

movie_directors = movie_directors[movie_directors["Directors"] != "Not specified"]
movie_directors = movie_directors.groupby(["Directors"]).size().reset_index(name = "Total Movies")
```

```python
movie_directors
```

```python
top_directors = pd.DataFrame(movie_directors.sort_values(by=["Total Movies"],ascending= False)).head(10)
top_directors
```

```python
#visualising the top movie directors

sns.barplot(y = "Directors", x = "Total Movies", data = top_directors, palette = "pastel")
plt.title("Top Movie Directors")
plt.show()
```

```python
#visualising the top movie directors

sns.barplot(y = "Directors", x = "Total Movies", data = top_directors, palette = "pastel")
plt.title("Top Movie Directors")
plt.show()
```

```python
plt.figure(figsize = (10,8))

ax = sns.barplot(
    y="Directors",
    x="Total Movies",
    data=top_directors,
    palette="pastel")
ax.bar_label(ax.containers[0], padding= 3)
plt.title("Top Movie Directors", fontsize = 16, pad = 15)
plt.xlabel("Total Movies")
plt.ylabel("Directors")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,8))

ax = sns.barplot(
    y="Directors",
    x="Total Movies",
    data=top_directors,
    palette="pastel")

ax.bar_label(ax.containers[4], padding= 7.5)

plt.title("Top Movie Directors", fontsize = 16, pad = 15)
plt.xlabel("Total Movies")
plt.ylabel("Directors")
plt.tight_layout()
plt.show()
```

```python
#expanding the countries to only include one country

countries = netflix_movie["country"].str.split(',', expand=True).stack()

#removing the white space before and after country names to avoid duplicate

countries = countries.str.strip('')
```

```python
#top movie countries

top_countries = pd.DataFrame(top_countries)

#renaming the column

top_countries.columns = ["Movies count"]
top_countries
```

```python
#A barplot to show the top countries by movies

plt.figure(figsize = (17,10))
sns.barplot(x = top_countries.index, y = top_countries["Movies count"], data = top_countries, color = "royalblue")
plt.xlabel("Countries")
plt.title("Top movie producing countries")
```

**Top movie actors**

```python
actors = netflix_movie["cast"].str.split(",", expand=True).stack()
actors = pd.DataFrame(actors)
actors.head()
```

```python
actors.columns = ["cast"]
```

```python
actors = actors[actors["cast"] != "Not listed"]
actors.head()
```

```python
#selecting the top 10 actors

top_actors = actors.value_counts().head(10)
top_actors
```

```python
#creating a dataframe

top_actors = pd.DataFrame(top_actors)

#renaming the column

top_actors.columns = ["Movies acted"]
top_actors
```

```python
#barchart to show the top actor movies

top_actors.plot(kind = "barh")
plt.xlabel("Count of movies")
plt.ylabel("Actors")
plt.title("top actors");
```
